**Usage Example**

```python
import PyOcclusion
```

All the user interactions with PyOcclusion are performed through creating an instance of VideoEditor class

```python
editor = PyOcclusion.VideoEditor(20, 0, 17, 17, 500, height=1080, width=1920)
```

edit() allows creating a new video file using parameters specified while instantiating the class

```python
editor.edit('video.mp4', "newVideo.mp4")
```

showOcclusion() allows user to preview the expected outlook of the occlusion to be applied to the video file

```python
editor.showOcclusion("picture.jpg", (12,8))
```
